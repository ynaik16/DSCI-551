##### working on products data

```python
temp = pd.DataFrame()
lists = {}
for i in range(400): 
    print(i)
    if os.path.isfile(products_directory+'/'+products[i]):
        with open(products_directory+'/'+products[i]) as f:
            data = json.load(f)
            dfs = pd.json_normalize(data['items'])
            lists[i] = dfs
    else:
        continue
        
```

```python
temp=pd.concat(lists, ignore_index = True)
```

```python
#temp.to_csv('raw_dataframe.csv')
```

```python
temp
```

```python
SELECTED_COLUMNS = ['itemId', 'name', 'msrp', 'salePrice','categoryPath','shortDescription', 'brandName', 'standardShipRate','bestMarketplacePrice','stock','numReviews']
```

```python
def clean_product_data(df):                 # cleaning the Df, droppiing NaNs
    global SELECTED_COLUMNS

    all = list((df.columns))                       # store all columns 

    for item in SELECTED_COLUMNS:
    #print(item)
        if item in all:
            all.remove(item)   

    new_df = df.drop(columns=all)  
    new_df.dropna(axis=0, inplace=True)    
    new_df = new_df.reset_index() 
    new_df = new_df.drop(columns=['index'])

    return new_df
```

```python
returned_df = clean_product_data(temp)
```

####### Q. Which 20 in-stock products(itemId) have the largest discount?

```python
items =[]
discounts = []
```

```python
for i in range(len(returned_df)):
    if re.search("available",returned_df['stock'][i], re.IGNORECASE):
        #print(returned_df['itemId'][i])
        discount = int(returned_df['msrp'][i]) - int(returned_df['salePrice'][i])
        items.append(returned_df['itemId'][i])
        discounts.append(discount)
        
d = {'items': items, 'dis': discounts}
max_discount = pd.DataFrame(d)
max_discount
```

```python
answer = max_discount.sort_values(['dis'], ascending=False)
```
